# Leetcode 23

Given an array of linked-lists lists, each linked list is sorted in ascending order.

Merge all the linked-lists into one sort linked-list and return it

## Thoughts
I still had to have assistance from other sources, which is slightly disapointing. I managed to code a decent solution, which was apparently too inefficient. I tried to merge each item together one at a time, but it blew up in time (became O (N) ^ 2 as I was constantly remerging items)  

The second approach required some research, where I decided to just take the minimal item out of the list (seems easy enough, I just was overhtinking the whole affair) and then commit the next item to end of list every time  
Time: O(K\*N)  
Space: O(1) (Given result and inputs do not take extra space)  

Third approach would be placing and constructing a priority queue to pop off and requeue with the next element, making for an Nlog(k) solution

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        
        head = None
        for i in range(len(lists)):
            pointList1 = head
            pointList2 = lists[i]
            
            res = None
            while pointList1 and pointList2:
                if pointList1.val < pointList2.val:
                    if not res:
                        res = ListNode(pointList1.val)
                        head = res
                    else:
                        res.next = ListNode(val = pointList1.val)
                        res= res.next

                    pointList1 = pointList1.next
                else:
                    if not res:
                        res = ListNode(pointList2.val)
                        head = res
                    else:
                        res.next = ListNode(pointList2.val)
                        res= res.next
                    pointList2 = pointList2.next
            if pointList1:
                if not res:
                    res = pointList1
                    head = res

                else:
                    res.next = pointList1
            elif pointList2:
                if not res:
                    res = pointList2
                    head = res
                else:
                    res.next = pointList2
        return head
        
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        
        head = res = None
        
        lists = [i for i in lists if i!= None]
        
        while len(lists) != 0:
            minEle = ListNode(math.inf)
            pos = -1
            for i, j in enumerate(lists):
                if j.val < minEle.val:
                    minEle = j
                    pos = i
            lists[pos] = lists[pos].next
            if lists[pos] == None:
                lists.pop(pos)
            if res:
                res.next = ListNode(minEle.val)
                res = res.next
            else:
                head = res = ListNode(minEle.val)
        return head
        
```
