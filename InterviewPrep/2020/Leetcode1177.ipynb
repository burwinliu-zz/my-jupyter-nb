# Leetcode 1177
## Problem
Given a string s, we make queries on substrings of s.

For each query queries[i] = [left, right, k], we may rearrange the substring s[left], ..., s[right], and then choose up to k of them to replace with any lowercase English letter. 

If the substring is possible to be a palindrome string after the operations above, the result of the query is true. Otherwise, the result is false.

Return an array answer[], where answer[i] is the result of the i-th query queries[i].

Note that: Each letter is counted individually for replacement so if for example s[left..right] = "aaa", and k = 2, we can only replace two of the letters.  (Also, note that the initial string s is never modified by any query.)

## Solution
### Solution 1
#### Thoughts
Had to be helped along a bit for optimization. Instead of dicts for each substring (see solution 2) do one pass to determine the substrings and then use those substrings calculated to determine the number of charecters at each position to be rearranged, which would save time of needing to reitterate over substrings which cna potentially be the entire length of the string. As such, the calculation goes from O(N) per query to O(26) -> O(1)  

#### Pseudo Code
Store substring combos for [0, i] for all i from [0 to n] into an array that will act as a representation of substring from 0 to i where each occurence of a letter is 
    stored at the ith position as one extra occurence  
Then, for each query, calculate the substring by subtracting the two positions in the store (store[j] - store[i] for each element)    
Determine out how many non "paired off" items there are   
for each non "paired off" pairs, calculate as one replacement  
if no more replacements remain, it is false; else for all pairs and replacements remain/no replacements needed, append true  
return result  
#### Analysis
Space : O(N) (for n is length of str, where 26 length array for each item in str)  
Time: O(M) (for M is number of queries as 26 operations to determine number of mismatches, plus a constant amount to determine if there are mismatches)

### Solution 2
#### Thoughts
Initial Naive solution -- using a dictionary, for each query, count nmumber of chars through each substirng, and determine if can be rearranged and reduced to palindrome with replacements.

#### Analysis
Space: O(N^N) where N is the length of the string  
Time: O(M ^ N) where M is the number of queries and N is the length of the string

```python
from collections import defaultdict

class Solution:
    def canMakePaliQueries(self, s: str, queries: List[List[int]]) -> List[bool]:
        charToInt = {
            'a': 1,
            'b': 2,
            'c': 3,
            'd': 4,
            'e': 5,
            'f': 6,
            'g': 7,
            'h': 8,
            'i': 9,
            'j': 10,
            'k': 11,
            'l': 12,
            'm': 13,
            'n': 14,
            'o': 15,
            'p': 16,
            'q': 17,
            'r': 18,
            's': 19,
            't': 20,
            'u': 21,
            'v': 22,
            'w': 23,
            'x': 24,
            'y': 25,
            'z': 26
        }
        result = []
        store = []
        current = [0 for i in range(26)]
        for i in s:
            current[charToInt[i] - 1] += 1
            store.append(current.copy())
        for i, j, k in queries:
            err = k * 2
            if (j - i + 1) % 2 != 0:
                err += 1
            if i == 0:
                iArr = [0 for i in range(26)]
            else:
                iArr = store[i - 1]
            jArr = store[j]
            for i in range(26):
                if (iArr[i] - jArr[i] )%2 != 0:
                    if err == 0:
                        result.append(False)
                        break
                    err -= 1
            else:
                result.append(True)
            
        return result
```
