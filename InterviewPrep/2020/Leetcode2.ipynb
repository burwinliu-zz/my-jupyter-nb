# Leetcode 2
## Problem
You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order and each of their nodes contain a single digit. Add the two numbers and return it as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.
## Solution
### Solution 1
#### Thoughts
Basic elementary linked list with two pointers
#### Analysis
Space and Time: O(N) -- storing O (N) items for the result and O(N) Items must be traversed to sum up
### Solution 2
Improvement on Solution 1 with slightly more efficient version
### Solution 3
Doing this via recursion!

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        
        carry = 0
        tail = head = ListNode()
        p1, p2 = l1, l2
        while p1 or p2:
            tail.val = carry
            if p1:
                tail.val += p1.val
                p1 = p1.next
            if p2:
                tail.val += p2.val
                p2 = p2.next
            carry = int((tail.val) / 10)
            tail.val = (tail.val) % 10
            if not p1 and not p2:
                if carry != 0:
                    tail.next = ListNode(carry)
                return head
            tail.next = ListNode()
            tail = tail.next
        if carry != 0:
            tail.next = ListNode(carry)
        return head
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        
        carry = 0
        tail = head = backtrack = ListNode()
        p1, p2 = l1, l2
        while p1 or p2:
            tail.val = carry
            if p1:
                tail.val += p1.val
                p1 = p1.next
            if p2:
                tail.val += p2.val
                p2 = p2.next
            carry = int((tail.val) / 10)
            tail.val = (tail.val) % 10
            tail.next = ListNode()
            backtrack = tail
            tail = tail.next
        backtrack.next = None
        if carry != 0:
            backtrack.next = ListNode(carry)
        return head
            
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        def helper(l1, l2, carry):
            next1, next2 = None, None
            if not l1 and not l2:
                if carry == 0:
                    return None
                return ListNode(carry)
            total = carry
            if l1:
                total += l1.val
                next1 = l1.next
            if l2: 
                total += l2.val
                next2 = l2.next
            
            return ListNode(total%10, helper(next1, next2, int(total/10)))
        
        return helper(l1, l2, 0)
            
```
