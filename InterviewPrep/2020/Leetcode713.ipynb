## Problem
Your are given an array of positive integers nums.

Count and print the number of (contiguous) subarrays where the product of all the elements in the subarray is less than k.

else, increment p1 and if p2 lags behind, update p2 with the total (tracking subarrays sum  
### Solution 2
Simple optimization by realizing that the reasoning behind one edge case was because I wasn't incremenint to len(nums)

```python
class Solution:
    def numSubarrayProductLessThanK(self, nums: List[int], k: int) -> int:
        # Add the distance between the two pointers as the new subarrays.
        # I.E [a, b, c] 0, 0 -- is less then, distance 0. Therefore + 1
        # 0, 1 -- distance 1, therefore + 2 -- 1 for 0, 1; 1 for 1,1
        # then 0, 2 --distance 2, therefore + 3 -- 1 for 0, 2; 1, 2; 2, 2
        # Then is 1, 2 is reached, ignroe until p2 moves
        # If 1,3 then add 1, 3; 2, 3; 3, 3 --
        # aka distance + 1 of each movement. go until p1 == len(nums -1)
        
        # Questions to ask
        #   Edge case -- what happens if length of array is 0
        #       Assumption -- return 0
        #   Forgot to ask -- edge case, what happens if no items Subarrays found **
        
        p1, p2 = 0, 0
        numSub = 0 # Number of subarrays 
        if len(nums) == 0: # Edge case
            return 0
        total = nums[p1] 
        while p1 != len(nums) - 1:
            if total < k:
                numSub += 1 + (p2 - p1)
                if p2 != len(nums)-1:
                    p2 += 1
                    total *= nums[p2]
                else:
                    return numSub
                
            else:

                total /= nums[p1]
                p1 += 1
                if p1 == len(nums) - 1:
                    if total < k:
                        numSub += 1 + (p2 - p1)
                if p2 < p1:
                    p2 += 1
                    total *= nums[p2]
                    

        return numSub

            
```
