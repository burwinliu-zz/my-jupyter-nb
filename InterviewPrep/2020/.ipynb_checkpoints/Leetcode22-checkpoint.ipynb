# Leetcode 22
## Problem
Given n pairs of parentheses, write a function to generate all combinations of well-formed parentheses.  

For example, given n = 3, a solution set is:

## Thoughts
The paranthesis are generated by recursion -- and each subproblem can be broken down to is it possible to add a ( and if it is possible to add ). This is determined by the number of opened and closed parentheses -- and if its less then the max total (n). By adding all possible opening and closed parentheses in all subproblems where allowed, we can generate all possible combinations that are valid

## Analysis
Time: O(Exponential in nature) -- but hard to calculate  -- according to google, Catalan number?  
Space: O(Same as time), as at max the stack would generate all possibilities in the stack to be recorded

```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        res = []
        callStack = [("", 0, 0)]
        
        while len(callStack) != 0:
            cur, numOpen, numCompleted = callStack[-1]
            callStack = callStack[:-1]
            if numCompleted == n:
                res.append(cur)
                continue
            if numOpen != n and numOpen + numCompleted < n:
                callStack.append((cur + "(", numOpen + 1, numCompleted))
            if numOpen != 0:
                callStack.append((cur + ")", numOpen - 1, numCompleted + 1))
        return res
            
```
